# Step 1: Imports

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

# Step 2: Create Data

```python
x = torch.tensor(
    [[1.0],
     [2.0],
     [3.0],
     [4.0]]
)

y = torch.tensor(
    [[2.0],
     [4.0],
     [6.0],
     [8.0]]
)
```

# Step 3: Create Model

```python
model = nn.Linear(
    in_features=1,
    out_features=1
)
```

# Step 4: Loss Function

```python
loss_fn = nn.MSELoss()
```

# Step 5: Optimizer

```python
optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.01
)
```

# Step 6: Training Loop

```python
losses = []

for epoch in range(500):

    predictions = model(x)

    loss = loss_fn(
        predictions,
        y
    )

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    losses.append(loss.item())

    if epoch % 50 == 0:
        print(
            f"Epoch {epoch}, Loss: {loss.item():.4f}"
        )
```

# Step 7: Inspect Learned Parameters

```python
print("\nWeight:")
print(model.weight)

print("\nBias:")
print(model.bias)
```

# Step 8: Test

```python
test = torch.tensor([[10.0]])

prediction = model(test)

print("\nPrediction for 10:")
print(prediction)
```

# Step 9: Plot Loss

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.show()
```
