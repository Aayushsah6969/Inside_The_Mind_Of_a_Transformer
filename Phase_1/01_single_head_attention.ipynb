## Imports

```python
import torch
import math
```

## Create Example Q, K, V

```python
Q = torch.tensor([
    [1.0, 0.0],
    [0.0, 1.0],
    [1.0, 1.0]
])

K = torch.tensor([
    [1.0, 0.0],
    [1.0, 1.0],
    [0.0, 1.0]
])

V = torch.tensor([
    [10.0, 0.0],
    [0.0, 20.0],
    [5.0, 5.0]
])
```

## Print Shapes

```python
print("Q Shape:", Q.shape)
print("K Shape:", K.shape)
print("V Shape:", V.shape)
```

## Compute Raw Attention Scores

```python
scores = Q @ K.T

print(scores)
```

## Scale Scores

```python
d_k = K.shape[-1]

print(d_k)
```

```python
scaled_scores = scores / math.sqrt(d_k)

print(scaled_scores)
```

## Apply Softmax

```python
weights = torch.softmax(
    scaled_scores,
    dim=-1
)

print(weights)
```

## Verify Rows Sum To 1

```python
print(weights.sum(dim=-1))
```

## Use Value Vectors

```python
output = weights @ V

print(output)
```

## Full Attention Function

Now combine everything.

```python
def attention(Q, K, V):

    scores = Q @ K.T

    scaled_scores = scores / math.sqrt(
        K.shape[-1]
    )

    weights = torch.softmax(
        scaled_scores,
        dim=-1
    )

    output = weights @ V

    return output, weights
```

## Run Function

```python
output, weights = attention(
    Q,
    K,
    V
)

print("Attention Weights")
print(weights)

print("\nOutput")
print(output)
```
